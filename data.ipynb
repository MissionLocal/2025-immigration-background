```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('data/ACSDT5Y2023.B05002_2025-10-23T135826/ACSDT5Y2023.B05002-Data.csv')
```

```python
df
```

```python
# make column names lowercase

df.columns = df.columns.str.lower()
```

```python
# I want to remove the first characters from geoid 1400000US to just have the census tract number

df['geo_id'] = df['geo_id'].str.replace('1400000US', '')
```

```python
df
```

```python
# remove the last column 

df = df.iloc[:, :-1]
```

```python
# I want to keep the following columns in a df called df_reduced: B05002_001E, B05002_013E, B05002_014E, B05002_021E

df_reduced = df[['geo_id', 'b05002_001e', 'b05002_013e', 'b05002_014e', 'b05002_021e']]
```

```python
df_reduced
```

```python
# rename columns to: geo_id, total_population, foreign_born, foreign_born_naturalized, foreign_born_not_naturalized

df_reduced = df_reduced.rename(columns={
    'b05002_001e': 'total_population',
    'b05002_013e': 'foreign_born',
    'b05002_014e': 'foreign_born_naturalized',
    'b05002_021e': 'foreign_born_not_naturalized'
})

df_reduced
```

```python
# remove row at index 0 

df_reduced = df_reduced.drop(index=0)
```

```python
# make a percentage column of foreign_born over total_population called pct_foreign_born

df_reduced['pct_foreign_born'] = (df_reduced['foreign_born'].astype(float) / df_reduced['total_population'].astype(float)) * 100

df_reduced
```

```python
df_reduced.pct_foreign_born.mean()
```

```python
# do the same for pct_foreign_born_naturalized and pct_foreign_born_not_naturalized

df_reduced['pct_foreign_born_naturalized'] = (df_reduced['foreign_born_naturalized'].astype(float) / df_reduced['foreign_born'].astype(float)) * 100

df_reduced['pct_foreign_born_not_naturalized'] = (df_reduced['foreign_born_not_naturalized'].astype(float) / df_reduced['foreign_born'].astype(float)) * 100

df_reduced
```

```python
# round to one decimal place

df_reduced['pct_foreign_born'] = df_reduced['pct_foreign_born'].round(1)
df_reduced['pct_foreign_born_naturalized'] = df_reduced['pct_foreign_born_naturalized'].round(1)
df_reduced['pct_foreign_born_not_naturalized'] = df_reduced['pct_foreign_born_not_naturalized'].round(1)

df_reduced
```

```python
# read file census_bay_area_tracts_2020_8121511352956101895

gdf = gpd.read_file('data/census_bay_area_tracts_2020_8121511352956101895.geojson')
```

```python
gdf
```

```python
# find a row where the geoid is 06097154308

gdf[gdf['geoid'] == '06097154308']
```

```python
# join df_reduced to gdf on geo_id and geoid

gdf_merged = gdf.merge(df_reduced, left_on='geoid', right_on='geo_id', how='left')

gdf_merged
```

```python
# export as mapdata.geojson in the docs folder

gdf_merged.to_file('docs/mapdata.geojson', driver='GeoJSON')
```
