```python
import pandas as pd
import geopandas as gpd
```

```python
! ls
```

```python
df = pd.read_csv('data/ACSDT5Y2023.B05002_2025-10-23T135826/ACSDT5Y2023.B05002-Data.csv')
```

```python
df
```

```python
# make column names lowercase

df.columns = df.columns.str.lower()
```

```python
# I want to remove the first characters from geoid 1400000US to just have the census tract number

df['geo_id'] = df['geo_id'].str.replace('1400000US', '')
```

```python
df
```

```python
# remove the last column 

df = df.iloc[:, :-1]
```

```python
# I want to keep the following columns in a df called df_reduced: B05002_001E, B05002_013E, B05002_014E, B05002_021E

df_reduced = df[['geo_id', 'b05002_009e','b05002_001e', 'b05002_013e', 'b05002_014e', 'b05002_021e']]
```

```python
df_reduced
```

```python
# rename columns to: geo_id, total_population, foreign_born, foreign_born_naturalized, foreign_born_not_naturalized

df_reduced = df_reduced.rename(columns={
    'b05002_001e': 'total_population',
    'b05002_013e': 'foreign_born',
    'b05002_009e': 'foreign_born_native',
    'b05002_014e': 'foreign_born_naturalized',
    'b05002_021e': 'foreign_born_not_naturalized'
})

df_reduced
```

```python
# remove row at index 0 

df_reduced = df_reduced.drop(index=0)
```

```python
# make a percentage column of foreign_born over total_population called pct_foreign_born

df_reduced['pct_foreign_born'] = (df_reduced['foreign_born'].astype(float) / df_reduced['total_population'].astype(float)) * 100

df_reduced
```

```python
df_reduced.pct_foreign_born.mean()
```

```python
# do the same for pct_foreign_born_naturalized and pct_foreign_born_not_naturalized

df_reduced['pct_foreign_born_naturalized'] = (df_reduced['foreign_born_naturalized'].astype(float) / df_reduced['total_population'].astype(float)) * 100

df_reduced['pct_foreign_born_not_naturalized'] = (df_reduced['foreign_born_not_naturalized'].astype(float) / df_reduced['total_population'].astype(float)) * 100

df_reduced['pct_foreign_born_native'] = (df_reduced['foreign_born_native'].astype(float) / df_reduced['total_population'].astype(float)) * 100

df_reduced
```

```python
# show me the range of the pct_foreign_born column

df_reduced['pct_foreign_born'].min(), df_reduced['pct_foreign_born'].max()
```

```python
# round to one decimal place

df_reduced['pct_foreign_born'] = df_reduced['pct_foreign_born'].round(1)
df_reduced['pct_foreign_born_naturalized'] = df_reduced['pct_foreign_born_naturalized'].round(1)
df_reduced['pct_foreign_born_not_naturalized'] = df_reduced['pct_foreign_born_not_naturalized'].round(1)

df_reduced
```

```python
# make pct_foreign_born_native rounded to one decimal place

df_reduced['pct_foreign_born_native'] = df_reduced['pct_foreign_born_native'].round(1)
```

```python
# read file census_bay_area_tracts_2020_8121511352956101895

gdf = gpd.read_file('data/census_bay_area_tracts_2020_8121511352956101895.geojson')
```

```python
gdf
```

```python
# find a row where the geoid is 06097154308

gdf[gdf['geoid'] == '06097154308']
```

```python
# join df_reduced to gdf on geo_id and geoid

gdf_merged = gdf.merge(df_reduced, left_on='geoid', right_on='geo_id', how='left')

gdf_merged
```

```python
gdf_merged = gdf_merged[gdf_merged['total_population'].astype(float) >= 100]
```

```python
# export as mapdata.geojson in the docs folder

gdf_merged.to_file('docs/mapdata.geojson', driver='GeoJSON')
```

```python
# read Bay_Area_County_Polygons_20251024

gdf_counties = gpd.read_file('data/Bay_Area_County_Polygons_20251024.geojson')
```

```python
gdf_counties
```

```python
# do a spatial join to get county names into gdf_merged

gdf_final = gpd.sjoin(gdf_merged, gdf_counties[['county', 'geometry']], how='left', predicate='intersects')

gdf_final
```

```python
gdf_final
```

```python
# make total a float in gdf_final

gdf_final['total_population'] = gdf_final['total_population'].astype(float)
```

```python
population_by_county = gdf_final.groupby('county')['total_population'].sum().reset_index()
```

```python
population_by_county
```

```python
# do the same for foreign_born

gdf_final['foreign_born'] = gdf_final['foreign_born'].astype(float)

foreign_born_by_county = gdf_final.groupby('county')['foreign_born'].sum().reset_index()

foreign_born_by_county
```

```python
# join both dataframes on county

county_summary = population_by_county.merge(foreign_born_by_county, on='county')
```

```python
county_summary
```

```python
# calculate pct_foreign_born at county level

county_summary['pct_foreign_born'] = (county_summary['foreign_born'] / county_summary['total_population']) * 100
```

```python
county_summary
```

```python
# get a summer of not naturalized

gdf_final['foreign_born_not_naturalized'] = gdf_final['foreign_born_not_naturalized'].astype(float)

not_naturalized_by_county = gdf_final.groupby('county')['foreign_born_not_naturalized'].sum().reset_index()

county_summary = county_summary.merge(not_naturalized_by_county, on='county')

county_summary
```

```python
# get a pct of not naturalized at county level

county_summary['pct_foreign_born_not_naturalized'] = (county_summary['foreign_born_not_naturalized'] / county_summary['total_population']) * 100

county_summary
```

```python
gdf_final['foreign_born_naturalized'] = gdf_final['foreign_born_naturalized'].astype(float)

foreign_born_by_county = gdf_final.groupby('county')['foreign_born_naturalized'].sum().reset_index()

foreign_born_by_county
```

```python
# merged both dataframes on county

county_summary = county_summary.merge(foreign_born_by_county, on='county')
```

```python
county_summary
```

```python
# make a pct of naturalized at county level

county_summary['pct_foreign_born_naturalized'] = (county_summary['foreign_born_naturalized'] / county_summary['total_population']) * 100

county_summary
```
